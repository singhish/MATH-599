# Homework 1 Problem 2.5.4

## Problem Statement
Breast cancer is a pertinent problem in womens' health that is a prevalent cause of death across women of all age ranges. The properties of breast cancer can be better understood by examining various attributes of breast-cancer-caused tumor masses, and carrying out statistical analyses across many samples of these masses.

Wolberg et al. of the University of Wisconsin have curated a dataset that utilizes 569 samples, each consisting of 30 real-valued features, of breast cancer tumor mass measurements. Each sample is labeled based on its diagnosis: `M` if it is malignant, and `B` if it is benign.

As it is impossible to visualize this many features at once, PCA provides a robust method for reducing the dimensionality of this dataset to observe if there is a difference in the distributions of malignant tumor masses and benign tumor masses. Thus, the goal of this project is to successfully reduce the dimensionality of this dataset to 1, 2, and 3 dimensions using the PCA method prescribed in class. In the process, the encoding and decoding functions for each dimension (characterized by their matrices) are to be determined, as well.

## Methods
Wolberg et al. provide their data in the file `wdbc.data` (located at https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data). To successfully reduce the dimensionality of this data, we use the `pandas` Python library to split the data into two separate datasets, `malignant` and `benign`, and perform PCA on each dataset.

We first model each dataset as a matrix

$X \in \rm I\!R^{\textit{n} \times \textit{m}}$

where $n$ corresponds to the number of features of the dataset (here, 30), and $m$ being the number of samples (357 for the `benign` data and 212 for `malignant`).

Next, for each dataset, we form a normalized matrix

$X' \in \rm I\!R^{\textit{n} \times \textit{m}}$

that is the result of each component of each row vector in $X$ subtracted by the mean across all components of the row vector (i.e. the average value of the feature across all samples). This is done with the help of Python's `numpy` library.

Then, we perform SVD on each $X'$ using `numpy`'s `linalg.svd` function, resulting in three matrices:

$X' = V \Lambda U^T$, $V \in \rm I\!R^{\textit{n} \times \textit{s}}$, $\Lambda \in \rm I\!R^{\textit{s} \times \textit{s}}$, $U \in \rm I\!R^{\textit{m} \times \textit{s}}$

where $s$ corresponds to the number of singular values obtained by the decomposition.

For each encoding dimension of interest, we obtain a matrix $D \in \rm I\!R^{\textit{n} \times \textit{l}}, $ $1 \leq l \leq s$, consisting of the first $l$ column vectors of $V$, where $l$ corresponds to the encoding dimension. Here, we also calculate the recovery error of the encoding dimension $l$:

$re = \sum\limits_{i=l+1}^{s} \sigma_i^2$

where $\sigma_i$ denotes the $i^{\rm th}$ singular value along the diagonal of $\Lambda$.

Finally, we encode each dataset:

$C = D^T X'$

where $C \in \rm I\!R^{l \times m}$. For each dataset, we can plot each column vector of $C$ using Python's `matplotlib` library to assess our results visually and determine whether or not there is a difference between how the samples of `benign` are distributed compared to the samples of `malignant`.

It is worth noting that for each $D$, the respective encoding function $f$ and decoding function $g$ are defined as the following:

$f(\textbf{x}) = D^T \textbf{x}$

$g(\textbf{c}) = D \textbf{c}$

where $\textbf{x} \in \rm I\!R^\textit{n}$ is a column vector of $X'$ and $\textbf{c} \in \rm I\!R^\textit{l}$ is a column vector of $C$. The process of obtaining the matrix $C$ of every encoded column vector of each dataset was accelerated using the matrix multiplication between $D^T$ and $X'$ described above. However, it is just as possible to multiply $D^T$ by each individual $\textbf{x}$ to obtain each corresponding $\textbf{c}$.

## Analysis
From each plot, it is interesting to observe that PCA did indeed reveal a difference in the distribution of the measurements for malignant tumor masses and benign tumor masses. Visually, the former appear to be much more spread out and variable, whereas in the latter case, tumor masses appear to be clustered closer to the origin of the principal component axes. This appears to indiciate that malignant tumor masses show much more overt irregularities, whereas benign tumor masses tend toward a central standard of normalcy. However, this distinction is not clear-cut, as many malignant tumor masses occur on the fringe of the benign "blob" -- using larger values for $l$ may reveal further information about whether or a boundary between malignant masses and tumor masses exist.

The work done here has harbors implicitations for future use-cases of other machine learning models, both unsupervised and supervised, for the classification of tumor mass samples. For instance, a support vector machine could be used to determine a robust boundary between malignant and benign tumor mass samples, or a multilayer perceptron classifier could be trained on this data to identify whether or not new tumor mass samples are malignant. PCA offers a means of avoiding the "curse of dimensionality" to simplify the calculations involved with these models (such as $l_2$ loss).

````python
import pandas as pd

df = pd.read_csv('wdbc.data', names=[i for i in range(-1, 31)])\
       .rename(columns={-1: 'id', 0: 'diagnosis'})\
       .set_index('id')

df
````

````python
benign = df[df['diagnosis'] == 'B'][[c for c in df.columns if c != 'diagnosis']]

benign
````

````python
malignant = df[df['diagnosis'] == 'M'][[c for c in df.columns if c != 'diagnosis']]

malignant
````

````python
import numpy as np

############################################################
# The functions below define the steps of the PCA process. #
############################################################

# convert pandas dataframe into X  R^(n by m)
def to_X(data):
    return data.to_numpy().T

# normalize rows of X to create X'  R^(n by m)
def get_Xpr(X):
    n = X.shape[0]
    Xpr = np.array([X[i] - np.mean(X[i]) for i in range(n)])
    
    assert Xpr.shape == X.shape  # ensure X' and X are of the same shape
    assert all([np.isclose(0, np.mean(r)) for r in Xpr])  # ensure means are 0
    
    return Xpr

# perform SVD and return D matrix and recovery error
def get_D(Xpr, l):
    V, L, Ut = np.linalg.svd(Xpr, full_matrices=False)
    n = Xpr.shape[0]
    s = L.size
    
    assert V.shape == (n, s)  # ensure V  R^(n by s)
    
    D = V[:, :l]
    re = sum([sig ** 2 for i, sig in enumerate(L) if i >= l + 1])
    
    return D, re
````

````python
X = to_X(malignant)
Xpr = get_Xpr(X)

D1, re1_m = get_D(Xpr, 1)
D2, re2_m = get_D(Xpr, 2)
D3, re3_m = get_D(Xpr, 3)

print(f'l = 1 encoding matrix (malignant samples) (shape={D1.T.shape}): \n{D1.T}', end='\n\n')
print(f'l = 1 decoding matrix (malignant samples) (shape={D1.shape}): \n{D1}', end='\n\n')
print(f'l = 1 recovery error: {re1_m}')

# create C matrix consisting of column vectors c  R^l for the malignant samples @ l = 1
# this is the same as applying f(x) = D^T x to all x where x is a column vector of X'
# this step is repeated for l = 2, 3 for the malignant samples and l = 1, 2, 3 for benign samples below
mal_l1 = D1.T @ Xpr
````

````python
print(f'l = 2 encoding matrix (malignant samples) (shape={D2.T.shape}): \n{D2.T}', end='\n\n')
print(f'l = 2 decoding matrix (malignant samples) (shape={D2.shape}): \n{D2}', end='\n\n')
print(f'l = 2 recovery error: {re2_m}')

mal_l2 = D2.T @ Xpr
````

````python
print(f'l = 3 encoding matrix (malignant samples) (shape={D3.T.shape}): \n{D3.T}', end='\n\n')
print(f'l = 3 decoding matrix (malignant samples) (shape={D3.shape}): \n{D3}', end='\n\n')
print(f'l = 3 recovery error: {re3_m}')

mal_l3 = D3.T @ Xpr
````

````python
X = to_X(benign)
Xpr = get_Xpr(X)

D1, re1_b = get_D(Xpr, 1)
D2, re2_b = get_D(Xpr, 2)
D3, re3_b = get_D(Xpr, 3)

print(f'l = 1 encoding matrix (benign samples) (shape={D1.T.shape}): \n{D1.T}', end='\n\n')
print(f'l = 1 decoding matrix (benign samples) (shape={D1.shape}): \n{D1}', end='\n\n')
print(f'l = 1 recovery error: {re1_b}')

ben_l1 = D1.T @ Xpr
````

````python
print(f'l = 2 encoding matrix (benign samples) (shape={D2.T.shape}): \n{D2.T}', end='\n\n')
print(f'l = 2 decoding matrix (benign samples) (shape={D2.shape}): \n{D2}', end='\n\n')
print(f'l = 2 recovery error: {re2_b}')

ben_l2 = D2.T @ Xpr
````

````python
print(f'l = 3 encoding matrix (benign samples) (shape={D3.T.shape}): \n{D3.T}', end='\n\n')
print(f'l = 3 decoding matrix (benign samples) (shape={D3.shape}): \n{D3}', end='\n\n')
print(f'l = 3 recovery error: {re3_b}')

ben_l3 = D3.T @ Xpr
````
